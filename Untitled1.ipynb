```python
#libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

#Loading Data
titanic_df = pd.read_csv('C:/Users/HP/Downloads/Initial Online Test Dataset/titanic.csv')

#Encoding the Categorical Columns
titanic_df['Sex'] = titanic_df['Sex'].map({'male': 0, 'female': 1})

#Filling the missing values in 'Embarked' column and encode it
titanic_df['Embarked'].fillna('S', inplace=True)
titanic_df['Embarked'] = titanic_df['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})

#Filling the missing values in 'Age' with the median value
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)

#Plotting the Class Distributions
plt.figure(figsize=(10, 8))
sns.countplot(x='Survived', data=titanic_df)
plt.title("Distribution of Survival on Titanic")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Counter")
plt.show()

#Spliting the Data into predictive Features (X) and Target (y)
# Drop irrelevant columns
X = titanic_df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin', 'Survived'])
y = titanic_df['Survived']

#Splitting the Data into Train and Test Sets
X_train, X_test, y_train, y_test = train_test_split(X, y)

#Training the model
log_reg = LogisticRegression(max_iter=190)
log_reg.fit(X_train, y_train)

#Evaluating the model
y_pred = log_reg.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

#Displaying the accuracy and classification report
print(f"Accuracy Outcomes: {accuracy * 100:.2f}%")
print("The classification Report:\n", classification_rep)
```
